```python
# f.eid: id of a participant
# f.31.0.0 reported sex (phenotype)
# f.22001.0.0 genetic sex
# f.22009.0.1 - f.22009.0.5 top 3 principle components for genetic data
# f.22021.0.0 kinship

# the number in () is the length of this datafieds reported in the UKbiobank
# the n-m is actually the suffix of a datafield existed in UKB44409.tab file

# f.41200: Operative procedures - main OPCS4 (53)    0-48

# f.41202: Diagnoses - main ICD10   (75)    0-65
# f.41203: Diagnoses - main ICD9    (28)    0-27
# f.41204: Diagnoses - secondary ICD10 (184) 0-183
# f.41205: Diagnoses - secondary ICD9  (30)  0-29

# f.41210: Operative procedures - secondary OPCS4 (97) 0-85
# f.41270: Diagnoses - ICD10        (223)  0-212
# f.41271: Diagnoses - ICD9         (47)   0-46
# f.41272: Operative procedures - OPCS4  (124) 0-116
eid = "f.eid"
sex_n, sex_g = "f.31.0.0", "f.22001.0.0"
pcs = ["f.22009.0.{}".format(i) for i in range(1, 6)]

kinship = "f.22021.0.0"
ethnic = "f.21000.0.0"
main_opcs4s = ["f.41200.0.{}".format(i) for i in range(0, 49)]
main_icd10s = ["f.41202.0.{}".format(i) for i in range(0, 66)]
main_icd9s = ["f.41203.0.{}".format(i) for i in range(0, 28)]
second_icd10s = ["f.41204.0.{}".format(i) for i in range(0, 184)]
second_icd9s = ["f.41205.0.{}".format(i) for i in range(0, 30)]
second_opcs4s = ["f.41205.0.{}".format(i) for i in range(0, 86)]
icd10s = ["f.41270.0.{}".format(i) for i in range(0, 213)]
icd9s = ["f.41271.0.{}".format(i) for i in range(0, 47)]
opcs4s = ["f.41272.0.{}".format(i) for i in range(0, 117)]

data_fields = [eid, sex_n, sex_g, kinship, ethnic]
data_fields = data_fields + pcs + main_opcs4s + main_icd10s + main_icd9s\
            + second_icd10s + second_icd9s + second_opcs4s + icd10s + icd9s\
            + opcs4s

# the index in data_fields is not following the index in .tab file
print(len(data_fields))
# should be 830
```

```python
# remove any row that has NA data in the columns sex_n, sex_g, pcs, kinship,
# and ethnic
n_before_qc = df_origin.shape[0]
null_na_dfs = [sex_n, sex_g, kinship, ethnic] + pcs
df = df_origin.dropna(axis = 0, subset = null_na_dfs, inplace = False)
print("we have {} rows left after dropping {} rows with NA values in {} columns".format(
    df.shape[0], n_before_qc - df.shape[0], null_na_dfs))
df.head(10)
```

```python
# filter sex inconsistent data,for sex_n, sex_g columns:
# 1. either has NA will be dropped
# 2. sex_n != sex_g will be dropped

n_before_qc = df.shape[0]
df = df[(df[sex_n] == df[sex_g])]
print("After sex QC, we have {} samples left, {} removed".format(
    df.shape[0], n_before_qc - df.shape[0]))
df.head(10)
```

```python
# only keep samples with kinship == 0 
n_before_qc = df.shape[0]
df = df[(df[kinship] > -1) & (df[kinship] < 1)]
print("after kinship QC, we have {} samples left, {} removed".format(
    df.shape[0], n_before_qc - df.shape[0]))
df.head(10)
```

```python
def scatter_plot(pca_np, sorted_ethnic_labels, ethnic_label_map,
                 axis_indices = [0, 1, 2],
                 s = 0.2, 
                 alpha = 1,
                 center = None,
                 radius = None,
                 axis_limits = [None, None, None],
                 title = "Scatter Plot on Principle Components"
                ):
    """draw 4 subplots(3 2D and 1 3D) in one figure
    """
    n_rows, n_cols = 2, 2
    fig, axs = plt.subplots(n_rows, n_cols, figsize=(12,8))
    pca_np[:, -1] = pca_np[:, -1].astype(int)
    scatters, labels = [], []
    for i in range(n_rows):
        for j in range(n_cols):
            ax = axs[i,j]
            for ethnic_label in sorted_ethnic_labels:
                ethnic_key = get_key_by_label(ethnic_label_map, ethnic_label)
                data = pca_np[pca_np[:, -1] == ethnic_key]
                #print("ethnic_label: {}, category: {}, samples: {}".format(
                #    ethnic_label, data[0,6], data.shape))
                is2D = (i+j < 2)
                if is2D:# 2D
                    pc_x_index, pc_y_index = (i, j+1) if i == 0 else (i, 2)
                    pc_x = axis_indices[pc_x_index]
                    pc_y = axis_indices[pc_y_index]
                    scatter = ax.scatter(data[:, pc_x], data[:, pc_y], 
                           c = colors[ethnic_label],
                           s = s, 
                           alpha = alpha, 
                           label = ethnic_label,
                           edgecolors='none')
                    
                    if i+j == 0:
                        scatters.append(scatter)
                        labels.append(ethnic_label)
                    
                    ax.grid(True)
                    ax.set_xlabel(r'$PC_{}$'.format(pc_x+1), fontsize=10)
                    ax.set_ylabel(r'$PC_{}$'.format(pc_y+1), fontsize=10)
                    ax.set_title('Principle Components ({}-{})'.format(
                        pc_x+1, pc_y+1))
                    if center is not None and radius is not None:
                        circle1 = plt.Circle(center, radius, 
                                             color ='b',
                                             linestyle = '--',
                                             linewidth = 0.5,
                                             alpha = 0.5,
                                             fill=False)
                        ax.add_patch(circle1)

                    if axis_limits:
                        if(axis_limits[pc_x]):
                            ax.set_xlim(axis_limits[pc_x])
                        if(axis_limits[pc_y]):
                            ax.set_ylim(axis_limits[pc_y])
                    
                    
                else: # 3D
                    ax.axis("off")
                    ax.grid(False)
                    ax.legend(scatters, 
                       labels, 
                       scatterpoints=1,
                       loc = 'best', 
                       ncol=2,
                       bbox_to_anchor=(0.9, 0.9),
                       numpoints = 1,
                       scatteryoffsets = [0.375, 0.5, 0.3125],
                       markerscale = 20,
                       frameon = False,
                       fancybox = False,
                       framealpha = 1.0,
                       facecolor = "black",
                       title = "self reported ethnic background",
                       title_fontsize = 12,
                       borderpad = 0.2,
                       fontsize=8) 
                    pass
    
    fig.suptitle(title)
    fig.tight_layout(pad=1.08, w_pad=4, h_pad=1.0)
    plt.show()
    
    
```

```python
def get_key_by_label(ethnic_labels, label):
    for key in ethnic_labels.keys():
        if ethnic_labels[key] == label:
            return key
    
    
def scatter_plot_2D(pca_np, sorted_ethnic_labels, ethnic_label_map, 
                    pc_x = 0, pc_y = 1, s = 0.2, alpha = 1,
                    center = None,
                    radius = None,
                    xlim = None,
                    ylim = None
                   ):
    fig, ax = plt.subplots(figsize=(6,4))
    pca_np[:, -1] = pca_np[:, -1].astype(int)
    scatters, labels = [], []
    for ethnic_label in sorted_ethnic_labels:
        ethnic_key = get_key_by_label(ethnic_label_map, ethnic_label)
        data = pca_np[pca_np[:, -1] == ethnic_key]
        #print("ethnic_label: {}, category: {}, samples: {}".format(
        #    ethnic_label, data[0,6], data.shape))
        scatters.append(
            ax.scatter(data[:, pc_x], data[:, pc_y], 
                       c = colors[ethnic_label],
                       s = s, 
                       alpha = alpha, 
                       label = ethnic_label,
                       edgecolors='none'))
        labels.append(ethnic_label)

    if center is not None and radius is not None:
        circle1 = plt.Circle(center, radius, color='b', fill=False)
        ax.add_patch(circle1)
        
    if xlim:
        ax.set_xlim(xlim)
    if ylim:
        ax.set_ylim(ylim)
        
    ax.set_xlabel(r'$PC_{}$'.format(pc_x+1), fontsize=10)
    ax.set_ylabel(r'$PC_{}$'.format(pc_y+1), fontsize=10)
    ax.set_title('Principle Components ({}-{})'.format(pc_x+1, pc_y+1))
    
    
    fig.tight_layout()
    ax.legend(scatters,
              labels,
              scatterpoints=1,
              loc='best', 
              ncol=2,
              bbox_to_anchor=(1, 1),
              numpoints = 1,
              scatteryoffsets = [0.375, 0.5, 0.3125],
              markerscale = 4/s,
              frameon = False,
              fancybox = False,
              framealpha = 1.0,
              facecolor = "black",
              title = "self reported ethnic background",
              title_fontsize = 12,
              borderpad = 0.2,
              fontsize=8) 
    ax.grid(True)
    plt.show()
```

```python
def calculate_distance(df, center, col_name = "distance"):
    '''calculate the Euclidean distance of a sample in given dataframe to a 
    given 3D point in the 3D coordinates established by top 3 pcs, the new 
    distance will be added as the last column of the df
    params: 
        df: the dataframe, df should include columns of  pcs[0:3]
        center: a tuple indicating a point in 3D space, (x, y, z)
        thres_dis: a float number indicating the threshold distance, sample's
            distance larger than thres_dis will not be selected.
    returns:
        a list of eid indicating selected sampls.
    '''
    df[col_name] = df[pcs[0]].sub(center[0]).pow(2).add(
                   df[pcs[1]].sub(center[1]).pow(2).add(
                   df[pcs[2]].sub(center[2]).pow(2))).pow(0.5)
    return df

# test
#test_df = df.loc[0:10, pcs]
#result_df = calculate_distance(test_df, (0,0,0), 50)
#result_df.head(10)
#from math import sqrt
#manually calcualte the distance of first sample
#print(sqrt((-10.9392)**2 + 5.49112**2 +(-0.502779)**2))# = 12.250358
```
